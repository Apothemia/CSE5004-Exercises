## 3. BF Traversal


```python
def bfs(v, g, starting_node):  #fun
    queue.append(starting_node)
    v.append(starting_node)
    while queue:
        m = queue.pop(0)
        print(m, end=' ')
        for neighbour in g[m]:
            if neighbour not in v:
                v.append(neighbour)
                queue.append(neighbour)


graph = {
    'v1': ['v2', 'v8'],
    'v2': ['v1', 'v3', 'v4'],
    'v3': ['v2', 'v4', 'v5'],
    'v4': ['v3', 'v5', 'v6', 'v7'],
    'v5': ['v3', 'v4'],
    'v6': ['v4', 'v8'],
    'v7': ['v4'],
    'v8': ['v1', 'v6'],
}
visited = []
queue = []

print('Following is the Breadth-First Search')
bfs(visited, graph, starting_node='v5')
```

## 4. DF Traversal


```python
graph = {
    1: [2, 4, 5],
    2: [1, 3, 4],
    3: [2, 4, 6],
    4: [1, 2, 3, 5, 6, 7],
    5: [1, 4, 7, 8],
    6: [3, 4, 7, 10],
    7: [4, 5, 6, 8, 10],
    8: [5, 7, 9],
    9: [8, 10],
    10: [6, 7, 9]
}
visited = set()

def dfs(visited, graph, node):
    if node not in visited:
        print(node, end=' ')
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)


# Driver Code
print('Following is the Depth-First Search')
dfs(visited, graph, 5)
```

## 5. Topological Sort


```python
DAG = nx.DiGraph()
edges = [
    ('m', 'q'), ('m', 'r'), ('m', 'x'),
    ('n', 'o'), ('n', 'q'), ('n', 'u'),
    ('o', 'r'), ('o', 's'),
    ('p', 's'), ('p', 'o'), ('p', 'z'),
    ('q', 't'),
    ('r', 'u'), ('r', 'y'),
    ('s', 'r'),
    ('u', 't'),
    ('v', 'w'), ('v', 'x'),
    ('w', 'z'),
    ('y', 'v')
]
for edge in edges:
    DAG.add_edge(*edge)

nx.draw_planar(DAG, with_labels=True, node_color='lightblue')
```

```python
order = list(nx.topological_sort(DAG))
order
```

```python
orders = list(nx.all_topological_sorts(DAG))
len(orders)
```
