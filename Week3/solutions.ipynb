```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def hamilton(G):
    F = [(G, [list(G.nodes())[0]])]
    n = G.number_of_nodes()
    while F:
        graph, path = F.pop()
        confs = []
        neighbors = (node for node in graph.neighbors(path[-1])
                     if node != path[-1])
        for neighbor in neighbors:
            conf_p = path[:]
            conf_p.append(neighbor)
            conf_g = nx.Graph(graph)
            conf_g.remove_node(path[-1])
            confs.append((conf_g, conf_p))
        for g, p in confs:
            if len(p) == n:
                return p
            else:
                F.append((g, p))
    return None


G = nx.Graph()
edges = [
    ('R', 'W'), ('R', 'S'), ('R', 'Q'), ('W', 'V'), ('S', 'T'),
    ('V', 'T'), ('W', 'X'), ('X', 'H'), ('H', 'J'), ('J', 'V'),
    ('S', 'N'), ('N', 'M'), ('M', 'L'), ('L', 'T'), ('J', 'K'),
    ('K', 'L'), ('K', 'F'), ('F', 'G'), ('G', 'B'), ('B', 'C'),
    ('C', 'D'), ('D', 'F'), ('B', 'Z'), ('Z', 'Q'), ('Q', 'P'),
    ('P', 'C'), ('Z', 'X'), ('P', 'N'), ('G', 'H'), ('D', 'M')]
G.add_edges_from(edges)

nx.draw_networkx(G)
plt.show()
```

```python
hamilton(G)
```
